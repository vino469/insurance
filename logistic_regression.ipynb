```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('/home/intellect-1001/Downloads/insurance_data.csv')
print(data.head())
```

```python
print(data.info())
```

```python
X = data[['age']]
y = data['bought_insurance']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Predicted Values:")
print(y_pred)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
new_age = pd.DataFrame([[35]], columns=['age'])
prediction = model.predict(new_age)

if prediction[0] == 1:
    print("Person will buy insurance")
else:
    print("Person will not buy insurance")
```

```python
plt.scatter(data.age, data.bought_insurance)
plt.xlabel("Age")
plt.ylabel("Bought Insurance")
plt.title("Insurance Prediction")
plt.show() 
```
